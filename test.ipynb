```python
# Method and constructur of oop in python
class demo:
    a=19
    def sum(self):
        print(self.a)
    def difference(self,a,b):
        self.c=a-b
        print(self.c)  
    def __init__(self):
        self.b=self.a*self.a
        print(self.b)      
democlass=demo()
democlass.sum() 
democlass.difference(20,10)       
```

```python
# Example of simple inheritance in python
class a:
    def displaya(self):
        print("welcome to python a:")
class b(a):
    def displayb(self):
        print("welcome to python b:")

obj=b()
obj.displaya()
obj.displayb()
```

```python
# Example of multilevel inheritance in python
class a:
    def displaya(self):
        print("welcome to dsa a:")
class b(a):
    def displayb(self):
        print("welcome to dsa b:")
class c(b):
    def displayc(self):
        print("welcome to dsa c:")

obj=c()
obj.displaya()
obj.displayb()
obj.displayc()                        
```

```python
# Example of multiple inheritance in python
class a:
    def displaya(self):
        print("welcome to coding a:")
class b:
    def displayb(self):
        print("welcome to coding b:")
class c(a,b):
    def displayc(self):
        print("welcome to coding c:")
obj=c()
obj.displaya()
obj.displayb()
obj.displayc()                        
```

```python
# Example of encapsulation in oop of python
class student:
    def __init__(self):
        self.name=""
    def getname(self):
        return self.name
    def setname(self,name):
        self.name=name
obj=student()
obj.setname("Bibek sodari")
a=obj.getname()
print(a)            
```

```python
# example of function overriding in python
class student:
    def display(self):
        print("my school name is purwancal campus:")

class college(student):
    def display(self):
        super().display()
        print("my college is lies in dharan:") 

obj=college()
obj.display()
```

```python
# example of methon overloading in python
class area:
    def calculatearea(self,a=None,b=None):
        if a!=None and b!=None:
            print("area is:",a*b)
        elif a!=None:
            print("area is:",a*a)
        else:
            print("there is no area calculated:")

obj=area()
obj.calculatearea()
obj.calculatearea(10)
obj.calculatearea(10,20)                
```
